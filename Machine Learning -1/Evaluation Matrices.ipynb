## Accuracy

ratio between # of correctly classified points (both true positives and true negatives) and total points ==> correct number / total points

```python
Image(filename = PATH + 'accuracy.png',width=500, height=300)
```

for example the accuracy of above model = 11 (6 true positive and 5 true negatives)/14 = 78.57

__Note__ : accuracy is a bad measure for skewed data.
